```python
import numpy as np
import pandas as pd

df_patient=pd.read_csv("patient_details.csv")
df_medical=pd.read_csv("medical_records.csv")

df_patient.info()
df_medical.info()

df_patient.loc[df_patient['age']<0,'age']=np.nan
df_patient['age'].fillna(df_patient['age'].median(),inplace=True)
pd.set_option('display.max_rows', None)

df_patient['name'].fillna("Avaish",inplace=True)
df_patient.drop_duplicates(subset = 'patient_id', keep='first', inplace=True)

df_patient['gender']=df_patient['gender'].replace(
    {'m':'Male',
     'M':'Male',
     'f':'Female',
     'F':'Female'})

df_patient['city']=df_patient['city'].replace(
    {'mumbai':'Mumbai',
     'kolkata':'Kolkata'})

df_patient['height_cm']=pd.to_numeric(df_patient['height_cm'],errors='coerce')
df_patient.loc[df_patient['height_cm']<0,'height_cm']=np.nan
df_patient['height_cm'].fillna(df_patient['height_cm'].median(),inplace=True)

df_patient['admission_date'] = pd.to_datetime(df_patient['admission_date'],format="mixed")

df_medical.drop_duplicates(subset='patient_id', keep='first', inplace=True)
df_medical['sugar_level'].fillna(df_medical['sugar_level'].median(),inplace=True)
df_medical['bp']=df_medical['bp'].replace({'high':'140/90'})

#'index=False'
df_patient.to_csv('New_patient.csv',index=False)
df_medical.to_csv('New_medical.csv',index=False)

#merge=pd.merge(df_patient,df_medical,on='patient_id',how='outer')
#merge
```
