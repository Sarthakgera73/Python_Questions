```python
#Q.write the code for DFS
graph={'A':['B','C','D'],
       'B':['A','C','D'],
       'C':['A','B','D'],
       'D':['A','C','B']
      }
def dfs(start,goal,graph,visited=None):
    if visited is None:
        visited=set()
    print(start,end=" ")
    if start==goal:
        return True
    visited.add(start)
    for child in graph[start]:
        if child not in visited:
            if dfs(child,goal,graph,visited):
                return True
    return False

dfs('A','D',graph)
```
