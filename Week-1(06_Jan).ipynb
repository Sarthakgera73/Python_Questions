```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data={
    'Student':['Sarthak','Ansh','Vansh','Rudy','Arsh','Tanish','Arjun','Dev','Anuradha','Chirag','Rohit'],
    'Marks':[85,75,np.nan,91,87,59,71,94,77,63,82],
    'Age':[17,19,18,19,17,np.nan,18,17,16,19,np.nan],
    'Subject':['Maths','Chem','Phy','C_lab','Python','Hindi',np.nan,'English','Python','Maths','THF']
}
df=pd.DataFrame(data)
df
```

```python
#First we will fill the null values then perform basic lib operations on dataset.
df['Marks'].fillna(df['Marks'].median(),inplace=True)
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Subject'].fillna("AI-ML",inplace=True)
df
```

```python
#Bargraph between marks and subject.
plt.bar(df['Subject'],df['Marks'],color='red')
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.title("Marks Distribution by Subject")
plt.show()
```
