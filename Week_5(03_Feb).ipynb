```python
dic={}
n=int(input("Enter the number of node: "))
for i in range(n):
    a=input("Enter node name")
    dic[a]=[]
    h=int(input(f"Enter the number of neighbour: "))
    for j in range(h):
        temp=input("Enter neighbour name:")
        dic[a].append(temp)

print("Graph is: ")
print(dic)
```

```python
def dfs(v,goal,graph,visited=None):
    if visited is None:
        visited=set()
    print(v,end=" -> ")
    if v==goal:
        return True
    visited.add(v)
    for V in graph[v]:
        if V not in visited:
            if dfs(V,goal,graph,visited):
                return True

start=input("Enter start node :")
last=input("Enter Last node: ")
print("DFS Transversal: ")
dfs(start,last,dic)
```
